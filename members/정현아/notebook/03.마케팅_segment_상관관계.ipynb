### 상관관계 분석을 위한 전처리

```python
res['Segment'].value_counts()
```

```python
# Segment encoding
le = LabelEncoder()
res['Segment_en'] = le.fit_transform(res['Segment'])

# decoded = le.inverse_transform(encoded)
# print(decoded)  # ['red', 'blue', 'green', ...]
```

```python
res.head()
```

```python
drop_cols = ['기준년월', 'ID', 'Segment']
res_drop = res.drop(columns = drop_cols)
res_drop.head()
```

```python
corr_matrix = res_drop.corr()
```

```python
len(corr_matrix['Segment_en'][corr_matrix['Segment_en'].abs() >= 0.02])
```

```python
corr_list = corr_matrix['Segment_en'][corr_matrix['Segment_en'].abs() >= 0.02].index.tolist()[:-1]
```

```python
corr_list
```

```python
pd.concat([df1.iloc[:,:2], df1[corr_list]],axis=1).to_parquet(path1 + '마케팅_전처리_final.parquet')
```
