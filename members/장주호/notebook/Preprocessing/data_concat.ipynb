```python
# 기본
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 경고 뜨지 않게 설정
import warnings
warnings.filterwarnings('ignore')

# 그래프 설정
sns.set()



# 그래프 기본 설정
plt.rcParams['font.family'] = 'NanumGothic' #리눅스용
#plt.rcParams['font.family'] = 'Malgun Gothic'
# plt.rcParams['font.family'] = 'AppleGothic'
plt.rcParams['figure.figsize'] = 12, 6
plt.rcParams['font.size'] = 14
plt.rcParams['axes.unicode_minus'] = False

#input_file_dir = "/workspace/data/train/"
#output_file_dir = "/workspace/data/train/0.종합본/"
input_file_dir = "C:/Users/sherm/Desktop/workspace/likeLion_final_project/data/train/"
output_file_dir = "C:/Users/sherm/Desktop/workspace/likeLion_final_project//data/train/0.종합본/"
```

### 데이터 병합(docker환경에선 터짐)

```python
data_categories = {
    "회원정보": {"folder": "1.회원정보/", "suffix": "회원정보", "var_prefix": "customer"},
    "신용정보": {"folder": "2.신용정보/", "suffix": "신용정보", "var_prefix": "credit"},
    "승인매출정보": {"folder": "3.승인매출정보/", "suffix": "승인매출정보", "var_prefix": "sales"},
    "청구정보": {"folder": "4.청구입금정보/", "suffix": "청구정보", "var_prefix": "billing"},
    "잔액정보": {"folder": "5.잔액정보/", "suffix": "잔액정보", "var_prefix": "balance"},
    "채널정보": {"folder": "6.채널정보/", "suffix": "채널정보", "var_prefix": "channel"},
    "마케팅정보": {"folder": "7.마케팅정보/", "suffix": "마케팅정보", "var_prefix": "marketing"},
    "성과정보": {"folder": "8.성과정보/", "suffix": "성과정보", "var_prefix": "performance"}
}
```

```python
for cat in data_categories:
    df1 = pd.read_parquet(input_file_dir + data_categories[cat]['folder'] + f"201807_train_{data_categories[cat]['suffix']}.parquet")
    df2 = pd.read_parquet(input_file_dir + data_categories[cat]['folder'] + f"201808_train_{data_categories[cat]['suffix']}.parquet")
    df3 = pd.read_parquet(input_file_dir + data_categories[cat]['folder'] + f"201809_train_{data_categories[cat]['suffix']}.parquet")
    df4 = pd.read_parquet(input_file_dir + data_categories[cat]['folder'] + f"201810_train_{data_categories[cat]['suffix']}.parquet")
    df5 = pd.read_parquet(input_file_dir + data_categories[cat]['folder'] + f"201807_train_{data_categories[cat]['suffix']}.parquet")
    df6 = pd.read_parquet(input_file_dir + data_categories[cat]['folder'] + f"201807_train_{data_categories[cat]['suffix']}.parquet")
    dfs=[df1,df2,df3,df4,df5,df6]
    df_all=pd.concat(dfs,axis=0, ignore_index=True)
    print(df_all)
    df_all.to_parquet(output_file_dir+f"train_{data_categories[cat]['suffix']}.parquet")
    print(f"train_{data_categories[cat]['suffix']}.parquet" + '파일이 완성되었습니다~')
```
