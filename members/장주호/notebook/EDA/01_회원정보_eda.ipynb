📌 이 변수의 의미 (분석 관점)
이건 고객의 신용 상태나 위험도를 간접적으로 보여주는 신호 변수야:

많을수록: 다양한 카드가 있고 문제 없이 쓰고 있다는 뜻 → 신용 양호

적을수록: 카드가 정지되었거나 BL에 등재된 게 많다는 뜻 → 위험 신호

🎯 이 변수에서 파생시킬 수 있는 아이디어들
1️⃣ 전체 카드 수 대비 사용 가능 카드 비율
usable_ratio = usable_cards / total_valid_cards
→ 고객이 가진 카드 중 얼마나 ‘건강한 카드’ 비율이 높은가?

→ 위험 고객은 비율이 낮을 가능성 있음

2️⃣ BL 등재된 카드 수 자체
bl_cards = total_valid_cards - usable_cards
→ 명시적으로 BL 등재 카드 수를 변수로 추가해줄 수 있어

→ Segment E 같은 고위험군이 이 값이 높을 수도 있음
