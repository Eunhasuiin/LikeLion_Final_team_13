```python

all_cols = ex1.columns.tolist()
print("전체 컬럼 이름:")
print(all_cols)


str_cols = ex1.select_dtypes(include=['object']).columns.tolist()
print("\n문자형 컬럼 이름:")
print(str_cols)

import numpy as np
num_cols = ex1.select_dtypes(include=[np.number]).columns.tolist()
print("\n숫자형 컬럼 이름:")
print(num_cols)
```

```python
cols = ['가입통신회사코드', '거주시도명', '직장시도명', '_1순위신용체크구분', '_2순위신용체크구분', '연회비발생카드수_B0M', '상품관련면제카드수_B0M', '임직원면제카드수_B0M', '우수회원면제카드수_B0M', '기타면제카드수_B0M', 'Life_Stage']  # 실제 열 이름으로 바꿔주세요
print(ex1[cols].head(5))
```

```python
mapping = {
    'L사': 0,
    'K사': 1,
    'S사': 2
}

# map 후 결측(None)은 -1로 채우고 정수형으로 변환
ex1['가입통신회사코드'] = (
    ex1['가입통신회사코드']
      .map(mapping)        # L사→0, K사→1, S사→2, 그 외→NaN
      .fillna(-1)          # NaN→-1
      .astype(int)         # 정수형
)

print(ex1['가입통신회사코드'].value_counts(dropna=False))
```

```python
import pandas as pd
from pandas.api.types import CategoricalDtype

# 1) 17개 광역시도 리스트 (원하는 순서대로)
regions = [
    '서울', '부산', '대구', '인천', '광주', '대전', '울산', '세종',
    '경기', '강원', '충북', '충남', '전북', '전남', '경북', '경남', '제주'
]

# 2) 카테고리 타입 정의
region_type = CategoricalDtype(categories=regions, ordered=True)

# 3) 변환 적용
for col in ['거주시도명', '직장시도명']:
    ex1[col] = (
        ex1[col]
           .astype(str)                   # 혹시 숫자 타입 섞여 있어도 문자열로
           .astype(region_type)           # 카테고리로 변환
           .cat.codes                     # 0~16, 리스트 외 값은 -1
    )

print("거주시도명 코드 분포:", ex1['거주시도명'].value_counts(dropna=False).sort_index())
print("직장시도명 코드 분포:", ex1['직장시도명'].value_counts(dropna=False).sort_index())
```

```python
count_cols = [
    '연회비발생카드수_B0M',
    '상품관련면제카드수_B0M',
    '임직원면제카드수_B0M',
    '우수회원면제카드수_B0M',
    '기타면제카드수_B0M'
]

for col in count_cols:
    ex1[col] = (
        ex1[col]
          .astype(str)  # None → 'None', NaN → 'nan'
          .str.replace('개', '', regex=False)  # '0개' → '0', 'None' → 'None'
    )
    # 숫자로 변환, 에러는 NaN 처리 → NaN → -1 로 채우기 → 정수형
    ex1[col] = (
        pd.to_numeric(ex1[col], errors='coerce')
          .fillna(-1)
          .astype(int)
    )

```

```python
le_cols = ['_1순위신용체크구분', '_2순위신용체크구분', 'Life_Stage']

for col in le_cols:
    codes, uniques = pd.factorize(ex1[col], sort=True)
    ex1[col] = codes  # NaN은 -1, 나머지는 0,1,2… 순서대로
```

```python
import pandas as pd
import numpy as np
from statsmodels.stats.outliers_influence import variance_inflation_factor
from statsmodels.tools.tools import add_constant

num_df = ex1.select_dtypes(include=[np.number]).dropna()

X = add_constant(num_df)

vif_df = pd.DataFrame({
    'feature': X.columns,
    'VIF': [variance_inflation_factor(X.values, i)
            for i in range(X.shape[1])]
})

vif_df = vif_df[vif_df['feature'] != 'const'].reset_index(drop=True)
print(vif_df)
```

```python
high_vif = vif_df[vif_df['VIF'] > 10].reset_index(drop=True)

high_features = high_vif['feature'].tolist()
print(high_features)

print(high_vif)
```
