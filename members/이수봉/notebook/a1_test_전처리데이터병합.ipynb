```python
import pandas as pd
```

```python
df1 = pd.read_parquet('merge/01_test_회원정보.parquet')
df2 = pd.read_parquet('merge/02_test_신용정보.parquet')
df3 = pd.read_parquet('merge/03_test_승인매출정보.parquet')
df4 = pd.read_parquet('merge/04_test_청구입금정보.parquet')
df5 = pd.read_parquet('merge/05_test_잔액정보.parquet')
df7 = pd.read_parquet('merge/07_test_마케팅정보.parquet')
df8 = pd.read_parquet('merge/08_test_성과정보.parquet')

df3
```

```python
# 인덱스 맞추기 (행 순서가 같다는 전제)
df1 = df1.reset_index(drop=True)
df2 = df2.reset_index(drop=True)
df3 = df3.reset_index(drop=True)
df4 = df4.reset_index(drop=True)
df5 = df5.reset_index(drop=True)
df7 = df7.reset_index(drop=True)
df8 = df8.reset_index(drop=True)


# 열 기준으로 concat
df_concat = pd.concat([df1, df2, df3, df4, df5, df7, df8], axis=1)

# 결과 확인
print("합쳐진 데이터프레임 shape:", df_concat.shape)
print("컬럼 수:", df_concat.shape[1])
print("미리보기:")
print(df_concat.head())
```

```python
# 중복된 컬럼 이름 제거 (앞에서 등장한 것만 남기고 나머지 제거)
df_deduped = df_concat.loc[:, ~df_concat.columns.duplicated()]

print("중복 컬럼 제거 후 shape:", df_deduped.shape)
```

```python
cols = df_deduped.columns.tolist()
cols
```

```python
df_deduped.to_parquet('merge/result/001_test_전처리데이터병합.parquet')
```
