```python
import pandas as pd
from sklearn.cluster import KMeans
import numpy as np
```

```python
df = pd.read_parquet('result/train/04_train_승인매출정보_VIF제거.parquet')
```

```python
# 클러스터링에 사용할 수치형 컬럼만 선택 (예: ID 등 제외)
exclude_cols = ['ID']  # 제외할 컬럼 이름은 여기에 추가
features = df.select_dtypes(include=[np.number]).drop(columns=exclude_cols, errors='ignore')

# KMeans 클러스터링 (5개 클러스터)
kmeans = KMeans(n_clusters=5, random_state=42)
df['Segment'] = kmeans.fit_predict(features)

# 숫자 라벨을 문자(A~E)로 변환
label_map = {0: 'A', 1: 'B', 2: 'C', 3: 'D', 4: 'E'}
df['Segment'] = df['Segment'].map(label_map)

# 결과 확인
print(df['Segment'].value_counts())
```
